```python
#Train/Plane Crossing problem
import matplotlib.pyplot as plt

t=0                     # set time t=0
delta=5                 # time difference
S_train=200             # Position of train
S_Plane=0              # Position of Plane
distance=0              # distance between positions of traina nd plane


print('%4s %6s %6s' %('t','S-train', 'S-Plane'))

plt.plot

for i in range(100):
  t=t+delta                      # update time
  S_train=200+50*t/60            # position of train at any time
  S_Plane=300*t/60               # position of plane
  distance=S_Plane-S_train       # Error
  if (distance<0):               #train is ahead
     print('%4d %6.3f %6.3f' %(t,S_train,S_Plane))
  else:                          #train is ahead
     print('%4d %6.3f %6.3f' %(t,S_train,S_Plane))
     break

```
