fixed point iteration
$$ g(x)=1/2 \sqrt (10-x^3) $$

```python
def g(x):
  g=(1/2) * np.sqrt(10-x**3)
  return g
#============================
```

```python
import numpy as np
nit=10
a=1
b=2
xold=1.5
for i in range(nit):
  y1=g(xold)       # value on curve
  y2=y1            # Intercept it on starght line
  xnew=y2          #get new abcissa value_xnew=g(xold)
  err=np.abs(xold-xnew)
  xold=xnew
  print(i,xnew,err)
```
