```python
import numpy as np
def f(x):
  val=x**3+4*x**2-10
  return val
#==================
f(1), f(2)
```

```python
a=1
b=2
e=1e-2
w=b-a      #Width of interval

i=0
while (w>=e):
  p=(a+b)/2
  i=i+1
  if (f(a)*f(p)>0):
    a=p
  else:
    b=p
  #==================
  w=b-a
  print('width is',w)
#====================
#print('width is',w)
print('No. of ietrations performed',i)
print('Root is',b)



```

```python
def bisection(f,a,b,e):
  w=b-a      #Width of interval

  i=0
  while (w>=e):
    p=(a+b)/2
    i=i+1
    if (f(a)*f(p)>0):
      a=p
    else:
      b=p
    #==================
    w=b-a
    print('width is',w)
 #====================
  #print('width is',w)
  print('No. of ietrations performed',i)
  print('Root is',b)
  values=i,w,p
  return values
```
